## Visualizing Data

```python
num_rows = (len(df.columns) + 2) // 4
num_cols = min(len(df.columns), 4)

fix, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 4*num_rows))

axes = axes.flatten()

for i, feature in enumerate(df.columns[1:]):
    if i < len(axes):
        df[feature].hist(ax=axes[i])
        axes[i].set_title(feature)
        
plt.tight_layout()
plt.show()
```

```python
num_rows = (len(df.columns) + 2) // 4
num_cols = min(len(df.columns), 4)

fix, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 4*num_rows))

axes = axes.flatten()

for i, feature in enumerate(df.columns[1:-1]):
    if i < len(axes):
        sns.boxplot(x='smoking', y=feature, data=df, ax=axes[i])
        axes[i].set_title(feature)
        
plt.tight_layout()
plt.show()
```

## Partial Summarization

1. Target feature has ~60/40 distribution.
2. We have no sex feature, but we can clearly see by height, weight and waist greatly influences whether a person smokes or not.
3. Boxplot shows more clearly about data outliers.

## Examine 'biological' features on target in details.

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='age', palette='Spectral', hue='smoking', multiple='stack')

plt.xlabel("Age")
plt.ylabel("Density")

plt.legend(title="Smoking habits", labels=["Smoker", "Non-Smoker"])

plt.show()
```

## Summarization.

### Age
Basically, from 20 to 40 years old, the ratio of smokers is higher than that of non-smokers. The older a person is, the less likely he is smoking. In this conclusion, one can try to establish the theme that people with bad habits can have health problems, which can later lead to death at an earlier age.

### Height
In a graph of human height, you can see that there is a tendency for the ratio of smokers to non-smokers  to increase as a person's height increases. Here we can assume that most of the people who smoke are men, whose height, on average, is higher than that of women. Therefore, as height increases, the percentage of people who smokes also increases. But we cannot say it's clearly true since we have no sex feature in our data.

### Weight
In a graph of a person's weight, you can also notice that the percentage of people who smoke increases as the person's weight increases. Again, this can be attributed to the assumption that in the data the percentage of men who smoke is higher than women, and men weigh on average more than women due to their higher height.
